```python
# Extracting only the stories from the second column
sentences = df.iloc[:, 1].dropna().tolist()

# Initialize models
unigram_freq = Counter()  # Unigram model, counter is used for find the words and their frequencies
bigram_freq = defaultdict(Counter)  # Bigram model, defaultdict where each key is a word, and the value is a Counter that will count how many times each word appears after the key word.
trigram_freq = defaultdict(Counter)  # Trigram model, key is a tuple of two words (the first two words of a trigram), and the value is a Counter that counts the third word in the trigram.
starting_word_freq = Counter()

# Tokenizing sentences and creating models
for sentence in sentences:
    words = word_tokenize(sentence)  # Tokenize the sentence using NLTK
    if not words:
        continue  # Skip empty sentences to avoid error if the next sentences is empty and we are trying to get the first word

    # Update starting word frequency with the first word of the sentence
    starting_word_freq.update([words[0]])

    # update() method takes the entire list of words and increments their counts in the unigram_freq Counter
    unigram_freq.update(words)

    for i in range(len(words) - 1):
        bigram_freq[words[i]][words[i + 1]] += 1  # Bigram model; +=1 is used for showing the freq which is by adding the count of th word

    for i in range(len(words) - 2):
        trigram_freq[(words[i], words[i + 1])][words[i + 2]] += 1  # Trigram model; which captures the frequency of the words which are most likely to comes to after a fixed sequence of two words


# Select the top N most common starting words
topWords = 50
most_common_starting_words = starting_word_freq.most_common(topWords) #starting_word_freq is defined while tokenization

# Extract just the words from the most_common_starting_words tupe which has both words and their frequenices
starting_words = [word for word, freq in most_common_starting_words]

# Function to generate the next word based on the bigram model
def generate_next_word_bigram(previous_word):
    next_word_candidates = bigram_freq[previous_word]
    if next_word_candidates:
        # Select a word based on frequency; valeus represent the frequency of the word  in the corpus after the previous_words
        return random.choices(list(next_word_candidates.keys()), weights=list(next_word_candidates.values()))[0]
    else:
        # If no candidates, return the most frequent word that follows previous_word
        if previous_word in bigram_freq:
            most_common_word = bigram_freq[previous_word].most_common(1)
            if most_common_word:
                return most_common_word[0][0]  # Return the most common word
    return None

# Function to generate the next word based on the trigram model
def generate_next_word_trigram(first_word, second_word):
    next_word_candidates = trigram_freq[(first_word, second_word)]
    if next_word_candidates:

        #random.choices can return multiple items. Since we only need one next word,
        # we use [0] to access the first element of the list returned by random.choices
        return random.choices(list(next_word_candidates.keys()), weights=list(next_word_candidates.values()))[0]
    else:
        # If no candidates, return the most frequent word that follows the last two words
        if (first_word, second_word) in trigram_freq:
            most_common_word = trigram_freq[(first_word, second_word)].most_common(1)
            if most_common_word:
                return most_common_word[0][0]  # Return the most common word
    return None

# Function to generate a sentence using unigrams, bigrams, or trigrams
def generate_sentence(n):
    sentence_length = random.randint(5, 19)  # Random length between 5 and 19
    starting_word = random.choice(a)
    sentence = [starting_word]

    if n == 1:  # Unigram model
        for _ in range(sentence_length - 1):
            next_word = random.choice(list(unigram_freq.keys()))  # Choose a random word from the unigram model
            sentence.append(next_word)

    elif n == 2:  # Bigram model
        for _ in range(sentence_length - 1):
            next_word = generate_next_word_bigram(sentence[-1])  # Generate the next word, sentence[-1]; shows to start from the end of the sentence
            if next_word:
                sentence.append(next_word)
            else:
                break  # Exit if no next word is found

    elif n == 3:  # Trigram model
        while len(sentence) < sentence_length:
            if len(sentence) < 2:  # If we don't have enough words for trigram, go back to bigram
                next_word = generate_next_word_bigram(sentence[-1])
                if next_word:
                    sentence.append(next_word)
            else:
                next_word = generate_next_word_trigram(sentence[-2], sentence[-1])  # Generate the next word based on its frequeny
                if next_word:
                    sentence.append(next_word)
                else:
                    break  # Exit if no next word is found

    return ' '.join(sentence)

# Function to generate paragraphs
def generate_paragraphs(num_paragraphs=3, sentences_per_paragraph=5):
    paragraphs = []
    for _ in range(num_paragraphs):
        paragraph = []
        for _ in range(sentences_per_paragraph):
            n = 3 #it shows which model to use (uni,bi or tri)
            sentence = generate_sentence(n)  # Generate the sentence
            paragraph.append(sentence)
        paragraphs.append('\n'.join(paragraph))  # Join sentences with newline for the paragraph
    return '\n\n'.join(paragraphs)  # Join paragraphs with a double newline for spacing between paragraph
```
